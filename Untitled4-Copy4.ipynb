```python
## Divide the Data Into Training and Test Set
```

```python
dfX = myfd.drop (columns =['Class'])
sy = myfd.Class
sy
```

```python
from sklearn.model_selection import train_test_split
dfX_train, dfX_test, sy_train, sy_test = train_test_split(dfX,sy)
dfX_train, dfX_test, sy_train, sy_test
```

```python
### Preprocessing non-numerical classes with LabelEncoder()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le
```

```python
le = le.fit(sy_train)
le.classes_
```

```python
y_train = le.transform(sy_train)
sy_train, y_train
```

```python
### Preprocessing attributes with StandardScaler()
```

```python
dfX_train.to_numpy()
n1 = preprocessing.StandardScaler()
n1 = n1.fit(dfX_train.to_numpy())

X_train = n1.transform (dfX_train.to_numpy())
X_train
```

```python
## Building the machine learning model for classification
```

```python
X_train.shape
```

```python

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn = knn.fit(X_train,y_train)
knn
```

```python
## Evaluation of the model using test sets
```

```python
### Preprocess the datasets dfX_test and sy_test
```

```python
y_test = le.transform(sy_test.to_numpy())
sy_test, y_test
```

```python
X_test = n1.transform(dfX_test.to_numpy())
dfX_test, X_test
```

```python
### Apply the model to predict the class of X_test and compare with y_test
```

```python
knn.predict(X_test)
```

```python
y_test
```

```python
knn.score(X_test, y_test)
```

```python
knn.score(X_train, y_train)
```

```python
####CONFUSION MATRIX####
```

```python
y_true = knn.predict(X_test)
y_pred = y_test
sklearn.metrics.confusion_matrix(y_true, y_pred)
```
