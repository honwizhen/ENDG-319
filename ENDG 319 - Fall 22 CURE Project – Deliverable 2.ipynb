```python
sorted_df = df.sort_values('typeofcancer')
fig, axs = plt.subplots(figsize=(10, 2.5))
axs1 = plt.subplot2grid ( shape= (1,3), loc= (0,0) )
axs2 = plt.subplot2grid ( shape= (1,3), loc= (0,1) )
axs3 = plt.subplot2grid ( shape= (1,3), loc= (0,2) )
plt.tight_layout()

axs1.hist(sorted_df.iloc[0:212, 0], edgecolor='black', fc = 'none', label = 'c0 (Malignant)')
axs1.hist(sorted_df.iloc[212:569, 0], edgecolor='red', fc = 'none', label = 'c1 (Benign)')
axs1.set_xlabel('mean radius')
axs1.set_ylabel('frequency')
axs1.legend()

axs2.scatter(sorted_df.iloc[0:212, 1], sorted_df.iloc[0:212, 0], c='black', label = 'c0 (M)')
axs2.scatter(sorted_df.iloc[212:569, 1], sorted_df.iloc[212:569, 0], c='red', label = 'c1 (B)')
axs2.set_xlabel('mean perimeter')
axs2.set_ylabel('mean radius')
axs2.legend()

axs3.scatter(sorted_df.iloc[0:212, 2], sorted_df.iloc[0:212, 0], c='black', label = 'c0 (M)')
axs3.scatter(sorted_df.iloc[212:569, 2], sorted_df.iloc[212:569, 0], c='red', label = 'c1 (B)')
axs3.set_xlabel('mean area')
axs3.set_ylabel('mean radius')
axs3.legend()
```

    In the histogram we can see that there is a higher frequency of benign cancer compared to the malignant variety. Furthermore, the mean perimeter shows a linear progression with the malignant type of cancer having a superior value, in contrast, the mean area has an exponential progression though like the perimeter the malignant cancer has a larger value area on average.

It is evident that there are higher frequency of Benign cancer while Malignant variant has bigger mean radius. Fig 2 indicates that Malignant has bigger perimeter and radius thatn Benign. Lastly, the Belingn variant tend to have smaller area. In conclusion, Benign tends to be smaller that Malignant.
